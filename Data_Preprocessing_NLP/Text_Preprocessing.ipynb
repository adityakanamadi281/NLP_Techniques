```python
df_reviews = pd.DataFrame(reviews, columns=["Reviews"])
df_reviews["Review_Lowercase"] = df_reviews["Reviews"].str.lower()
df_reviews
```

### Removing Punctuation and Emojis

```python
import re
df_reviews["Review_NoPunctEmoji"] = df_reviews["Review_Lowercase"].apply(lambda x: re.sub(r'[^\w\s]','',x))
```

```python
df_reviews
```

#### Remove Stop Words

```python
import nltk 
nltk.download("stopwords")
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))
df_reviews["Review_NoStopWords"]= df_reviews["Review_NoPunctEmoji"].apply(lambda x: " ".join(word for word in x.split() if word not in stop_words))
```

```python
df_reviews
```

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()
df_reviews["Review_Stemmed"] = df_reviews["Review_NoStopWords"].apply(lambda x: " ".join(ps.stem(word) for word in x.split()))
```
