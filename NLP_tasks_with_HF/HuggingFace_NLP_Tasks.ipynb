```python
from transformers import pipeline

classifier = pipeline(
    "sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english",
    device=-1  # CPU
)

text = "I really like to work with transformers library and LLM use cases."

result = classifier(text)

print(f"Text: {text}")
print(f"Predicted Sentiment: {result[0]['label']} with confidence {result[0]['score']:.4f}")
```
